```python
# practice using fixed latitudes and longitudes
import requests
import gmaps
from config import gkey
```

```python
# set parameters to find hotel in Paris
params = {
    "radius": 5000,
    "types": 'lodging',
    "key": gkey,
    "location": "48.8566, 2.3522"}

# use base URL
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# make request and get JSON data from search, note Hotel Duo is under "Name" under 'results' under hotels
hotels = requests.get(base_url, params=params).json()
hotels
```
